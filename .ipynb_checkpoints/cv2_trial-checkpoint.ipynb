```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('apo.jpg',1)
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
face = img[120:285,420:630]
```

```python
plt.imshow(face)
```

```python
resize1 = cv2.resize(img,None,fx = 0.6,fy = 0.6,interpolation =  cv2.INTER_LINEAR)
```

```python
plt.imshow(resize1)
```

```python
mask = cv2.inRange(img,(0,50,100),(100,150,225))
```

```python
plt.imshow(mask)
```

```python
scaling_factor = 1/225.0
image = np.float32(img)
image = image*scaling_factor
```

```python
plt.imshow(image)
```

```python
image = image * (1.0/scaling_factor)
image = np.uint8(image)
```

```python
brighthight = cv2.add(image, np.ones(image.shape,dtype='uint8')*50)
```

```python
plt.imshow(brighthight)
```

```python
contrastPercentage = 30

# Clip the values to [0,255] and change it back to uint8 for display
contrastImage = image * (1+contrastPercentage/100)
clippedContrastImage = np.clip(contrastImage, 0, 255)
contrastHighClippedUint8 = np.uint8(clippedContrastImage)
```

```python
plt.imshow(contrastHighClippedUint8)
```
